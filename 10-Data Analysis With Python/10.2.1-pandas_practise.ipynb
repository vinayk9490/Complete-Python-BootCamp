```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Vinay', 'Rahul', 'Sneha', 'Ankit'],
    'Age': [25, 30, 28, 22],
    'Salary': [50000, 60000, 55000, 40000]
}, index=['a', 'b', 'c', 'd'])

print(df)
```

```python
print(df.loc['a']) #prints the entire data related to index 'a'
print("\n")
print(df.loc['a':'b','Name':'Salary'])
```
