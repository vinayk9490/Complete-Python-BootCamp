```python
#creating multiple plots 
#concept of subplots
x = [1, 2, 3, 4, 5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(10,5)) #setting the figure size

plt.subplot(1,2,1) #1 row, 2 columns, 1st plot
plt.plot(x,y1, color='green', marker='o', linestyle='-', linewidth=1, markersize=8)

plt.subplot(1,2,2) #1 row, 2 columns, 1st plot
plt.plot(y1,x, color='green', marker='o', linestyle='-', linewidth=1, markersize=8)
```

## 📊 Subplot grid explained: rows and columns

Great news: you’re very close to understanding this clearly, and I can make it simple in one shot.  
`plt.subplot(nrows, ncols, index)` is just a grid definition where you place your plots.

- `nrows` = number of horizontal lines of subplots (vertical stacking)
- `ncols` = number of vertical lines of subplots (horizontal stacking)
- `index` = which cell in that grid you’re currently drawing into (1-based count left→right, top→down)

### Example 1: 1 row, 2 columns
`plt.subplot(1, 2, 1)` and `plt.subplot(1, 2, 2)`

- grid:
  - [plot1][plot2]
- row count 1 → one horizontal strip
- col count 2 → two side-by-side plots

### Example 2: 2 rows, 1 column
`plt.subplot(2, 1, 1)` and `plt.subplot(2, 1, 2)`

- grid:
  - [plot1]
  - [plot2]
- row count 2 → two vertical blocks (stacked)
- col count 1 → one column only

### Example 3: 2 rows, 3 columns
`plt.subplot(2, 3, 4)` etc.

- grid:
  - [1][2][3]
  - [4][5][6]
- `index=4` is the first cell of second row

---

### 💡 Key idea
Rows and columns are like seats in a theatre:
- Rows = horizontal levels.
- Columns = vertical slots in each row.
- Index = which seat you occupy to draw that plot.

```python
#Bar Plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 20]

plt.bar(categories, values, color='grey')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
```

CREATING A SCATTER PLOT

```python
x = [1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y, color='purple', marker='x')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Scatter Plot Example')
```

PIE CHARTS WITH SAMPLE DATA

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, shadow=True)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.title('Pie Chart Example')
plt.show()
```
