```python
class Car:

    def __init__(self, model, color, year, engine):
        self.model = model
        self.color = color
        self.year = year
        self.engine = engine

    def engine_specs(self):
        print(f"The car belongs to {self.engine} engine family.")
```

```python
class Aura(Car):

    def __init__(self, model, color, year, engine, is_sedan):
        super().__init__(model, color, year, engine)
        self.is_sedan = is_sedan
    
    def aura_information(self):
        if (self.is_sedan==True):
            print(f"The car is a sedan: {self.is_sedan}")
        else:
            print(f"The car is not a sedan: {self.is_sedan}")
    
```

```python
variant_one = Aura("Hyundai Aura","Teel Green",2025,"Kappa",True)
variant_one.engine_specs()
variant_one.aura_information()
```

```python
class Cat:

    def __init__(self, cat_speak):
        self.cat_speak = cat_speak

class Dog:
    
    def __init__(self, dog_speak):
        self.dog_speak = dog_speak

class Animal(Cat, Dog):

    def __init__(self, cat_speak, dog_speak):
        Cat.__init__(self, cat_speak)
        Dog.__init__(self, dog_speak)
    
    def animal_speak(self):
        print(f"The cat says: {self.cat_speak}")
        print(f"The dog says: {self.dog_speak}")
```

```python
behaviour = Animal("Meow", "Woof")
behaviour.animal_speak()
```
